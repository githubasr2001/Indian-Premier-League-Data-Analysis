Let's create a function to merging old and new team names

```python
def change_teamname(x):
    if x=='Deccan Chargers':
        return 'Sunrisers Hyderabad'
    elif x=='Pune Warriors' or x=='Rising Pune Supergiant':
        return 'Rising Pune Supergiants'
    elif x=='Delhi Daredevils':
        return 'Delhi Capitals'
    elif x=='Kings XI Punjab':
        return 'Punjab Kings'
    else: 
        return x
```

```python
deliveries['batting_team']=deliveries['batting_team'].apply(change_teamname)
deliveries['bowling_team']=deliveries['batting_team'].apply(change_teamname)
```

```python
deliveries['batting_team'].unique()
```

```python
xlables=list(deliveries['batting_team'].unique())
plt.figure(figsize=(15,15))
plt.xticks(rotation = 90)
plt.grid(visible=True)
plt.title('Boundary percentage of each team',size=20,fontweight='bold')
for i in range(0,len(xlables)):
    plt.subplot(4,4,i+1)
    deliveries[deliveries['batting_team']==xlables[i]]['runs_off_bat'].value_counts().plot.pie(autopct='%.2f',radius=1.2)
    plt.xlabel(xlables[i],fontweight='bold')
    plt.ylabel('Runs')
plt.suptitle('Boundary Percentage',size=20,weight='bold')
```

```python
def runs_scored_by_team(batting_team):
    t=deliveries[deliveries['batting_team']==batting_team]
    return t['runs_off_bat'].sum()
    
    

    
    
```

```python
pm=list(deliveries['batting_team'].unique())
for i in range(0,len(pm)):
    print("Total Runs Scored by"+" "+pm[i]+" "+"in IPL 2008-2022"+":"+ str(runs_scored_by_team(pm[i])))
```

```python
def face_off(batsman):
    t1=deliveries[deliveries['striker']==batsman]
    return t1.groupby('bowler')['runs_off_bat'].sum().sort_values(ascending=False).head(10).plot.bar()
```
